```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('submission_0802.csv')
```

```python
df.head()
```

```python
len(df['EncodedPixels'])
```

```python
count = 0
for i in range(len(df['EncodedPixels'])):
    p = df['EncodedPixels'][i]
    if p == p:
        array = p.split(" ")
        if len(array) % 2 != 0:
            a = p[:-1] + str(" ") + p[-1]
            df['EncodedPixels'][i] = a
            count += 1
```

```python
count
```

```python
df.to_csv('submission_0802_1.csv', header=True, index=False)
```
